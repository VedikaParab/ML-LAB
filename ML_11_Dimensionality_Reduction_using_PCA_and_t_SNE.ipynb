```python
# Dimensionality Reduction using PCA and t-SNE

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Step 1: Load dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Step 2: Standardize the data
X_scaled = StandardScaler().fit_transform(X)

# Step 3: PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 4: t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200)
X_tsne = tsne.fit_transform(X_scaled)

# Step 5: Plot PCA result
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
for i, target in enumerate(np.unique(y)):
    plt.scatter(X_pca[y == target, 0], X_pca[y == target, 1], label=target_names[i])
plt.title("PCA Result (2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()

# Step 6: Plot t-SNE result
plt.subplot(1,2,2)
for i, target in enumerate(np.unique(y)):
    plt.scatter(X_tsne[y == target, 0], X_tsne[y == target, 1], label=target_names[i])
plt.title("t-SNE Result (2D)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend()

plt.show()
```
